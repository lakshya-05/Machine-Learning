```python
df = df.dropna(subset=["margin_low"])
df
```

```python
df.info()
```

```python
from sklearn.model_selection import train_test_split

x_train, x_test, y_train, y_test = train_test_split(df.drop(columns=["is_genuine"], axis=1), df.is_genuine, test_size=0.3)
```

```python
from sklearn.preprocessing import StandardScaler

scaler = StandardScaler()
scaled_x_train = scaler.fit_transform(x_train)

scaled_x_test = scaler.fit_transform(x_test)
```

```python
scaled_x_train.shape
```

```python
from sklearn.neighbors import KNeighborsClassifier

clsf = KNeighborsClassifier(n_neighbors=5)
clsf.fit(scaled_x_train, y_train)
```

```python
from sklearn.metrics import accuracy_score

y_pred = clsf.predict(x_test)

accuracy_score(y_test, y_pred)
```
