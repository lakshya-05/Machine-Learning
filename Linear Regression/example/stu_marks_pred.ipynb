```python
import pandas as pd
```

```python
df = pd.read_csv("Student_Marks.csv")
df.head()
```

```python
df = df.drop(columns=['number_courses'])
```

```python
import matplotlib.pyplot as plt

plt.scatter(df.Marks, df.time_study)
```

```python
from sklearn.linear_model import LinearRegression
from sklearn.model_selection import train_test_split
```

```python
X_train, X_test, y_train, y_test = train_test_split(df[["time_study"]], df["Marks"], test_size=0.3)
```

```python
lr = LinearRegression()
lr.fit(X_train, y_train)
```

```python
lr.predict([[3]])
```

```python
from sklearn.metrics import mean_squared_error

mean_squared_error(lr.predict(X_test), y_test)
```

```python
import numpy as np

y_pred = lr.predict(X_test)
plt.scatter(df.time_study, df.Marks)

idx = np.argsort(X_test.values.flatten())

plt.plot(
    X_test.values.flatten()[idx],
    y_pred[idx]
)

plt.xlabel("Time Study")
plt.ylabel("Marks")
```
