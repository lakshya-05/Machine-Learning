```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
def loss(x):
    return x**2
```

```python
def grad(x):
    return 2*x
```

```python
x = 10
lr = 0.1
```

```python
x_history = [x]
for _ in range(20):
    x = x - lr * grad(x)
    x_history.append(x)
```

```python
plt.plot(x_history, label="x values")
plt.axhline(0, color='gray', linestyle='--')
plt.title("Gradient Descent on y = x²")
plt.legend()
plt.show()
```
