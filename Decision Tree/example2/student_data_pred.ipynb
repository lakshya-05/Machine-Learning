```python
import pandas as pd
```

```python
df = pd.read_csv("StudentPerformanceFactors.csv")
df.head()
```

```python
df.columns
```

```python
df = df.drop(columns=['Parental_Involvement', 'Access_to_Resources','Internet_Access',
       'Tutoring_Sessions', 'Family_Income', 'Teacher_Quality', 'School_Type',
       'Peer_Influence', 'Physical_Activity', 'Learning_Disabilities',
       'Parental_Education_Level', 'Distance_from_Home'], axis=1)
```

```python
df.head()
```

```python
from sklearn.preprocessing import LabelEncoder
```

```python
le = LabelEncoder()
```

```python
df["Extracurricular_Activities"] = le.fit_transform(df["Extracurricular_Activities"])
df.head()
```

```python
df["Motivation_Level"] = le.fit_transform(df["Motivation_Level"])
df.head()
```

```python
df["Gender"] = le.fit_transform(df["Gender"])
df.head()
```

```python
from sklearn.model_selection import train_test_split

X_train, X_test, y_train, y_test = train_test_split(df.drop(columns=["Exam_Score"], axis=1), df["Exam_Score"], test_size=0.3)
```

```python
from sklearn.tree import DecisionTreeRegressor
```

```python
tree2 = DecisionTreeRegressor(max_depth=2)

tree2.fit(X_train, y_train)

y_pred2_train = tree2.predict(X_train)
y_pred2_test = tree2.predict(X_test)
```

```python
tree20 = DecisionTreeRegressor(max_depth=20)

tree20.fit(X_train, y_train)

y_pred20_train = tree20.predict(X_train)
y_pred20_test = tree20.predict(X_test)
```

```python
from sklearn.metrics import mean_squared_error, r2_score
```

```python
mse2_train = mean_squared_error(y_train, y_pred2_train)
mse2_test = mean_squared_error(y_test, y_pred2_test)

r2_train = r2_score(y_train, y_pred2_train)
r2_test = r2_score(y_test, y_pred2_test)
```

```python
mse20_train = mean_squared_error(y_train, y_pred20_train)
mse20_test = mean_squared_error(y_test, y_pred20_test)

r20_train = r2_score(y_train, y_pred20_train)
r20_test = r2_score(y_test, y_pred20_test)
```

```python
print(f"""
      max_depth = 2
Training: 
    MSE -> {mse2_train}
    r2  -> {r2_train*100}
Testing:
    MSE -> {mse2_test}
    r2  -> {r2_test*100}

    
      max_depth = 20
Training: 
    MSE -> {mse20_train}
    r2  -> {r20_train*100}
Testing:
    MSE -> {mse20_test}
    r2  -> {r20_test*100}
""")
```

### max_depth = 2 -> little bit underfitting
### max_depth = 20 -> overfitting
