```python
import numpy as np
import pandas as pd
from sklearn.datasets import load_iris
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LogisticRegression
from sklearn.ensemble import RandomForestClassifier
from sklearn.svm import SVC
from sklearn.model_selection import StratifiedKFold
from sklearn.model_selection import cross_val_score
```

```python
data = load_iris()
```

```python
dir(data)
```

```python
X = data.data
y = data.target
```

```python
X[0]
```

```python
y[0]
```

```python
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.2)
```

```python
model_lr = LogisticRegression()
model_lr.fit(X_train, y_train)
model_lr.score(X_test, y_test)
```

```python
model_rfc = RandomForestClassifier()
model_rfc.fit(X_train, y_train)
model_rfc.score(X_test, y_test)
```

```python
model_svm = SVC()
model_svm.fit(X_train, y_train)
model_svm.score(X_test, y_test)
```

```python
def get_score(model, X_train, y_train, X_test, y_test):
    model.fit(X_train, y_train)
    return model.score(X_test, y_test)
```

```python
Kfold = StratifiedKFold(n_splits=4)
```

```python
score_logistic = []
score_rfc = []
score_svm = []
```

```python
for train_index, test_index in Kfold.split(X, y):
    X_train = data.data[train_index]
    X_test = data.data[test_index]

    y_train = data.target[train_index]
    y_test = data.target[test_index]

    score_logistic.append(get_score(LogisticRegression(), X_train, y_train, X_test, y_test))
    score_rfc.append(get_score(RandomForestClassifier(), X_train, y_train, X_test, y_test))
    score_svm.append(get_score(SVC(), X_train, y_train, X_test, y_test))
```

```python
score_logistic
```

```python
score_rfc
```

```python
score_svm
```

```python
# cross_val_score does the same thing as above but in a more efficient way
scores_logistic1 = cross_val_score(LogisticRegression(), X, y, cv=Kfold)
scores_rfc1 = cross_val_score(RandomForestClassifier(), X, y, cv=Kfold)
scores_svm1 = cross_val_score(SVC(), X, y, cv=Kfold)
```
