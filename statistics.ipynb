```python
import numpy as np
from scipy import stats
```

```python
data = np.random.normal(10, 2, size=1000)

print("Mean:", np.mean(data))
print("Std Dev:", np.std(data))
```

```python
sample = np.random.choice(data, size=50)
t_stat, p_val = stats.ttest_1samp(sample, popmean=10)
print("T-stat:", t_stat, "P-value:", p_val)
```

```python
conf_int = stats.norm.interval(0.95, loc=np.mean(sample), scale=np.std(sample)/np.sqrt(len(sample)))
print("95% CI:", conf_int)
```
