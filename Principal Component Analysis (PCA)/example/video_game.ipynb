```python
import pandas as pd
```

```python
df = pd.read_csv("Video Games Sales (1980-2024) - Raw.csv")
df.head()
```

```python
df.info()
```

```python
df.shape
```

```python
df.columns
```

```python
df = df.drop(columns=['img', 'publisher', 'developer',
       'na_sales', 'jp_sales', 'pal_sales',
       'other_sales', 'release_date', 'last_update'], axis=1)
```

```python
df.head()
```

```python
df["title"].unique()
```

```python
df.shape
```

```python
df = df.dropna(subset=['critic_score'])
```

```python
df.shape
```

```python
df = df.drop_duplicates(subset='title')
```

```python
df.shape
```

```python
df.head()
```

```python
df = pd.get_dummies(df, columns=['genre', 'console'])
```

```python
df.describe()
```

```python
df = df.drop(columns=["title"], axis=1)
df.head()
```

```python
df.shape
```

```python
df.info()
```

```python
df['total_sales'] = df['total_sales'].fillna(df['total_sales'].median())
```

```python
X = df.drop(columns=["critic_score"], axis=1)
y = df["critic_score"]
```

```python
X.shape
```

```python
from sklearn.decomposition import PCA

pca = PCA(n_components=2)
X_pca = pca.fit_transform(X)
```

```python
X_pca.shape
```

```python
from sklearn.model_selection import train_test_split

X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3)
X_train_pca, X_test_pca, y_train_pca, y_test_pca = train_test_split(X_pca, y, test_size=0.3)
```

```python
df.corr(numeric_only=True)
```

```python
from sklearn.ensemble import RandomForestRegressor

model1 = RandomForestRegressor()
model1.fit(X_train, y_train)
```

```python
model2 = RandomForestRegressor()
model2.fit(X_train_pca, y_train_pca)
```

```python
from sklearn.metrics import r2_score
```

```python
pred1 = model1.predict(X_test)

r2_score(y_test, pred1)
```

```python
pred2 = model2.predict(X_test_pca)

r2_score(y_test_pca, pred2)
```
