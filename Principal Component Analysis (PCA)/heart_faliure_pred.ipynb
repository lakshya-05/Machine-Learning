```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn.model_selection import train_test_split
from sklearn.decomposition import PCA
```

```python
df = pd.read_csv('data/heart.csv')
df
```

```python
from sklearn.preprocessing import LabelEncoder

le = LabelEncoder()
df['Sex'] = le.fit_transform(df['Sex'])
df['ChestPainType'] = le.fit_transform(df['ChestPainType'])
df['RestingECG'] = le.fit_transform(df['RestingECG'])
df['ExerciseAngina'] = le.fit_transform(df['ExerciseAngina'])
df['Oldpeak'] = le.fit_transform(df['Oldpeak'])
df['ST_Slope'] = le.fit_transform(df['ST_Slope'])

df
```

```python
df.info()
```

No more categorical values

```python
from sklearn.preprocessing import StandardScaler

sc = StandardScaler()
scaled_data = sc.fit_transform(df.drop(columns=['HeartDisease'], axis=1))
```

```python
data = pd.DataFrame(scaled_data, columns=df.columns[:-1])
data
```

```python
X_train, X_test, y_train, y_test = train_test_split(data, df.HeartDisease, test_size=0.2, random_state=40)
```

```python
from sklearn.linear_model import LogisticRegression

lr = LogisticRegression(C=1)
lr.fit(X_train, y_train)
lr.score(X_test, y_test)
```

```python
from sklearn.svm import SVC

svm = SVC(kernel='linear', C=1)
svm.fit(X_train, y_train)
svm.score(X_test, y_test)
```

```python
from sklearn.ensemble import RandomForestClassifier

rf = RandomForestClassifier(max_depth=4)
rf.fit(X_train, y_train)
rf.score(X_test, y_test)
```

Let's go with Logistic Regression and can also go with SVC or Random Forest Classifier as all are giving similar accuracy scores

```python
from sklearn.decomposition import PCA

pca = PCA(0.90)
X_pca = pca.fit_transform(data)
X_pca.shape
```

```python
X_train_pca, X_test_pca, y_train, y_test = train_test_split(X_pca, df.HeartDisease, test_size=0.2, random_state=40)
```

```python
# going with logistic regression

lr1 = LogisticRegression(C=1)
lr1.fit(X_train_pca, y_train)
lr1.score(X_test_pca, y_test)
```
