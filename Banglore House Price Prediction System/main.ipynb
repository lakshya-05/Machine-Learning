```python
import numpy as np
import pandas as pd
```

```python
df = pd.read_csv('data/Bengaluru_House_Data.csv')
df.head()
```

### Data Cleaning

```python
df = df.drop(columns=['area_type', 'availability', 'society'])
df.head()
```

```python
df.shape
```

```python
df.isnull().sum()
```

```python
df.balcony = df.balcony.fillna(0)
df.head()
```

```python
df.bath = df.bath.fillna(df.bath.median())
df.head()
```

```python
df.isnull().sum()
```

```python
df = df.dropna()
df.isnull().sum()
```

```python
df.shape
```

```python
df['size'].unique()
```

```python
df['bhk'] = df["size"].apply(lambda x: int(x.split(" ")[0]))
df.head()
```

```python
def is_float(x):
    try:
        float(x)
    except:
        return False
    return True
```

```python
df[~df['total_sqft'].apply(is_float)]
```

```python
def convert_sqft_to_num(x):
    tokens = x.split('-')
    if len(tokens) == 2:
        return (float(tokens[0]) + float(tokens[1]))/2
    try:
        return float(x)
    except:
        return None
```

```python
convert_sqft_to_num('2200 - 4000')
```

```python
df["total_sqft"] = df["total_sqft"].apply(convert_sqft_to_num)
df.head()
```

```python
df.loc[30]
```

```python
df[~df['total_sqft'].apply(is_float)]
```

```python
df.head()
```

```python
df.location = df.location.apply(lambda x: x.strip())
len(df.location.unique())
```

```python
loc_vals = df['location'].value_counts()
loc_vals
```

```python
len(loc_vals[loc_vals <= 15])
```

```python
loc_vals_less_than_15 = loc_vals[loc_vals <= 15]
loc_vals_less_than_15
```

```python
df.location = df.location.apply(lambda x: 'other' if x in loc_vals_less_than_15 else x)
len(df.location.unique())
```

```python
df.head(10)
```

```python
df['price_per_sqft'] = df['price'] * 100000 / df['total_sqft']
df.head(10)
```

### Outlier Removal

```python
df[df.total_sqft/df.bhk < 300].head()
```

```python
df = df[~(df.total_sqft/df.bhk < 300)]
df.shape
```
