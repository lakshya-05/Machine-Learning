```python
import numpy as np
import pandas as pd
```

```python
df = pd.read_csv('data/Bengaluru_House_Data.csv')
df.head()
```

```python
df = df.drop(columns=['area_type', 'availability', 'society'])
df.head()
```

```python
df.shape
```

```python
df.isnull().sum()
```

```python
df.balcony = df.balcony.fillna(0)
df.head()
```

```python
df.bath = df.bath.fillna(df.bath.median())
df.head()
```

```python
df.isnull().sum()
```

```python
df = df.dropna()
df.isnull().sum()
```

```python
df.shape
```

```python
df['size'].unique()
```

```python
df['bhk'] = df["size"].apply(lambda x: int(x.split(" ")[0]))
df.head()
```

```python
def is_float(x):
    try:
        float(x)
    except:
        return False
    return True
```

```python
df[~df['total_sqft'].apply(is_float)]
```

```python
def convert_sqft_to_num(x):
    tokens = x.split('-')
    if len(tokens) == 2:
        return (float(tokens[0]) + float(tokens[1]))/2
    try:
        return float(x)
    except:
        return None
```

```python
convert_sqft_to_num('2200 - 4000')
```

```python
df["total_sqft"] = df["total_sqft"].apply(convert_sqft_to_num)
df.head()
```

```python
df[~df['total_sqft'].apply(is_float)]
```
